```python
# Import packages
import numpy
import matplotlib.pyplot as plt
```

```python
# Read a sample image
sample_image = plt.imread("landscape.jpg")
```

```python
print(sample_image)
```

```python
sample_image
```

```python
# 3D array shape
sample_image.shape
```

```python
# Extract arrays for different channels
R = sample_image[:, :, 0]
G = sample_image[:, :, 1]
B = sample_image[:, :, 2]
```

```python
# Assuming R, G, and B are the extracted channels
# Flatten the arrays to create histograms
R_flat = R.flatten()
G_flat = G.flatten()
B_flat = B.flatten()

# Plot histograms
plt.figure(figsize=(12, 6))

# Red channel
plt.subplot(1, 3, 1)
plt.hist(R_flat, bins=256, color='red', alpha=0.7)
plt.title('Red Channel')

# Green channel
plt.subplot(1, 3, 2)
plt.hist(G_flat, bins=256, color='green', alpha=0.7)
plt.title('Green Channel')

# Blue channel
plt.subplot(1, 3, 3)
plt.hist(B_flat, bins=256, color='blue', alpha=0.7)
plt.title('Blue Channel')

plt.tight_layout()
plt.show()
```
